```
#importing essential libraries
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
import tpot
```

```
#loading data
data = load_boston()
```

```
#converting numpy arry into pandas dataframe
boston_housing = pd.DataFrame(data=data.data , columns=data.feature_names)
```

```
#assining dependent and independent variables
X = boston_housing
y = data.target
```

```
#splitting data into train and test
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```
#defining model and fitting model
regressor = tpot.TPOTRegressor(generations=10, population_size=100, random_state=42,cv=5,verbosity=2, scoring='r2')
regressor.fit(X_train,y_train)
```

```
#validating model using tpot
regressor.score(X_test,y_test)
```

```
#traditional evalution and prediction using scikit-learn
from sklearn.metrics import r2_score
y_pred = regressor.predict(X_test)
r2_score(y_pred, y_test)
```
